# Further Sales Analysis Analysis

```python
# Product prices
product_prices = {
    'メンズ・シャンプーM': 2500,
    'レディース・シャンプーL': 3000,
    'ラグジュアリ・シャンプーEx': 5000
}

# Calculate total sales for each product
invoice_data['product1_sales'] = invoice_data['quantity_1'] * product_prices['メンズ・シャンプーM']
invoice_data['product2_sales'] = invoice_data['quantity_2'] * product_prices['レディース・シャンプーL']
invoice_data['product3_sales'] = invoice_data['quantity_3'] * product_prices['ラグジュアリ・シャンプーEx']
invoice_data.head(1)
```

```python
# Count the total number of rows before dropping
total_rows = len(invoice_data)

# Convert invoice_date column to datetime type
invoice_data['invoice_date'] = pd.to_datetime(invoice_data['invoice_date'], errors='coerce')

# Drop rows with non-date data in invoice_date column
invoice_data_mod = invoice_data.dropna(subset=['invoice_date'])

# Count the number of rows after dropping
remaining_rows = len(invoice_data_mod)

# Count original number of invoices by franchises
total_rows_by_franchise = invoice_data.groupby('franchise_name').size()

# Count rows that remained after removing corrupted invoices by franchises
total_rows_by_franchise_mod = invoice_data_mod.groupby('franchise_name').size()

# Calculate the dropped rows by franchise
dropped_rows_by_franchise = total_rows_by_franchise.sub(total_rows_by_franchise_mod)

# Calculate the dropped percentage by franchise
dropped_percentage_by_franchise = total_rows_by_franchise.sub(total_rows_by_franchise_mod) / total_rows_by_franchise * 100

# Calculate the percentage of dropped data
dropped_percentage = ((total_rows - remaining_rows) / total_rows) * 100
print("Percentage of dropped data: {:.2f}%".format(dropped_percentage))

# Create a dataframe storing the data retention results
data_retention = pd.concat([total_rows_by_franchise, total_rows_by_franchise_mod,
                           dropped_rows_by_franchise, dropped_percentage_by_franchise],
                          axis=1)
data_retention = data_retention.reset_index(drop=False)
data_retention = data_retention.rename(columns = {0: 'row_count_original', 1: 'row_count_cleaned', 2: 'dropped_row_count', 3: 'data_lost_percentage'})
data_retention
```

```python
df = invoice_data

# Convert invoice_date column to datetime type
invoice_data['invoice_date'] = pd.to_datetime(invoice_data['invoice_date'], errors='coerce')

# Drop rows with non-date data in invoice_date column
invoice_data = invoice_data.dropna(subset=['invoice_date'])

# Extract calendar month and year from invoice_date
invoice_data['calendar_month'] = invoice_data['invoice_date'].dt.month
invoice_data['calendar_year'] = invoice_data['invoice_date'].dt.year

# Calculate annual sales for each row
invoice_data['total_sales'] = invoice_data['product1_sales'] + invoice_data['product2_sales'] + invoice_data['product3_sales']

# Calculate monthly sales for each franchise and year
monthly_sales = invoice_data.groupby(['franchise_name', 'calendar_year', 'calendar_month'])['total_sales'].sum()

# Calculate annual sales for each franchise and year
annual_sales = invoice_data.groupby(['franchise_name', 'calendar_year'])['total_sales'].sum()

# Calculate the number of invoices and average sales per invoice for each month
invoice_number = invoice_data.groupby(['franchise_name', 'calendar_year', 'calendar_month'])['invoice_date'].count()
invoice_avg = monthly_sales / invoice_number

# Create a new DataFrame for sales by franchise, year, and month
sales_by_franchise_year_month_df = pd.DataFrame({
    'Franchise Name': monthly_sales.index.get_level_values('franchise_name'),
    'Year': monthly_sales.index.get_level_values('calendar_year'),
    'Month': monthly_sales.index.get_level_values('calendar_month'),
    'Sales': monthly_sales.values,
    'Invoice Number': invoice_number.values,
    'Invoice Avg': invoice_avg.values
})

# Create a new DataFrame for sales by franchise and year
sales_by_franchise_year_df = pd.DataFrame({
    'Franchise Name': annual_sales.index.get_level_values('franchise_name'),
    'Year': annual_sales.index.get_level_values('calendar_year'),
    'Sales': annual_sales.values
})

# Add the "total" row to the DataFrame for sales by franchise, year, and month
sales_by_franchise_year_month_df.loc[len(sales_by_franchise_year_month_df)] = ['Total', '', '',
                                                                              sales_by_franchise_year_month_df['Sales'].sum(),
                                                                              sales_by_franchise_year_month_df['Invoice Number'].sum(),
                                                                              sales_by_franchise_year_month_df['Sales'].sum() / sales_by_franchise_year_month_df['Invoice Number'].sum()]

# Add the "total" row to the DataFrame for sales by franchise and year
sales_by_franchise_year_df.loc[len(sales_by_franchise_year_df)] = ['Total', '',
                                                                  sales_by_franchise_year_df['Sales'].sum()]

# Reset the index of the DataFrames
sales_by_franchise_year_month_df = sales_by_franchise_year_month_df.reset_index(drop=True)
sales_by_franchise_year_df = sales_by_franchise_year_df.reset_index(drop=True)
```
