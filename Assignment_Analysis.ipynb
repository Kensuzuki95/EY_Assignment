```python
# Group the data by 'municipality' and 'prefecture', and sum the 'quantity' columns
municipality_sales = invoice_data.groupby(['municipality', 'prefecture']).agg({'quantity_1': 'sum', 'quantity_2': 'sum', 'quantity_3': 'sum'}).reset_index()
municipality_sales['quantity_total'] = municipality_sales['quantity_1'] + municipality_sales['quantity_2'] + municipality_sales['quantity_3']

# Print the resulting dataset
municipality_sales.columns
```

```python
# Merge the two dataset using municipality as a key
sales = pd.merge(income_data_mod[['municipality', 'prefecture', 'income']],
                 municipality_sales[['municipality', 'quantity_1', 'quantity_2', 'quantity_3', 'quantity_total']],
                 how='left', on = 'municipality', suffixes = ('_left', '_right'))
sales = sales.fillna(0)
sales['income'] = sales['income'].astype(float)
sales
```
